### Restore Glamaholic
If Glamaholic ever goes down, and want access to your glamour plates, you can run this notebook to get a .txt file which lists your glamours.

```python
# Import Packages
from garlandtools import GarlandTools
import json
import time as time
import progressbar as pb
```

Define your paths to your glamaholic json, the output file location and where you wish to store the Garland Tool cache.

```python
# Path to Glamaholic .json file
glam_file       = open(glam_file)
glam_file      = 'Path/to/glamaholic.json'
# Output file path
output_file    = 'Path/to/output.txt'

# Cache location for GarlondTools API
Cache_location = "Path/to/cache"
```

Run the following to contact Garland tools

```python
# From scratch
# Import Packages
from garlandtools import GarlandTools
import json
import time as time
import progressbar as pb
# Path to Glamaholic .json file
glam = open(glam_file)

#------------------------------------
# Begin code
#------------------------------------
# Where we store the output glamour plate information
output_dict = dict()

# Necessary definitions
wait_time = 5  # Wait time in seconds between queries. Be sensible to the API.
# --------------------------------
# Activate Garlondtools API
api = GarlandTools(cache_location=Cache_location)

# Create Empty output list
Print_string = []

# Load the .json file
data = json.load(glam)

# Calculate number of glamour plates
Num_plates = len(data['Plates'])
print("Number of glamour plates: " + str(Num_plates))
print("")

# For loop to identify items with each glamour
for i in pb.progressbar(range(Num_plates)):
    Plate           = data['Plates'][i]
    Plate_name      = Plate['Name']
    Plate_items     = Plate['Items']

    print("Name of Glamour Plate: " + Plate_name)
    print("Glamour plate {0} out of {1}".format(i+1 ,Num_plates))
    print("Number of items: {0}".format(len(Plate_items)-1))

    #--------------------------------------------
    # Main Hand
    #--------------------------------------------

    print("Main Hand for plate " + Plate_name)
    try:
        Plate_Main_Hand = Plate_items['Main Hand']['ItemId']
        response = api.item(Plate_Main_Hand)
        if response.ok:
            item_json = response.json()
            Plate_Main_Hand_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Main_Hand_name = "None"
    print("Item retrieved as: " + Plate_Main_Hand_name)
    print("\n")
    #--------------------------------------------
    # Body 
    #--------------------------------------------
    print("Body for plate " + Plate_name)
    try:
        Plate_Body = Plate_items['Body']['ItemId']
        response = api.item(Plate_Body)
        if response.ok:
            item_json = response.json()
            Plate_Body_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Body_name = "None"
    print("Item retrieved as: " + Plate_Body_name)
    print("\n")

    #--------------------------------------------
    # Head
    #--------------------------------------------
    print("Head for plate " + Plate_name)
    try:
        Plate_Head = Plate_items['Head']['ItemId']
        response = api.item(Plate_Head)
        if response.ok:
            item_json = response.json()
            Plate_Head_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Head_name = "None"
    print("Item retrieved as: " + Plate_Head_name)
    print("\n")

    #--------------------------------------------
    # Hands
    #--------------------------------------------
    print("Hands for plate " + Plate_name)
    try:
        Plate_Hands = Plate_items['Hands']['ItemId']
        response = api.item(Plate_Hands)
        if response.ok:
            item_json = response.json()
            Plate_Hands_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Hands_name = "None"
    print("Item retrieved as: " + Plate_Hands_name)
    print("\n")

    #--------------------------------------------
    # Legs
    #--------------------------------------------
    print("Legs for plate " + Plate_name)
    try:
        Plate_Legs = Plate_items['Legs']['ItemId']
        response = api.item(Plate_Legs)
        if response.ok:
            item_json = response.json()
            Plate_Legs_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Legs_name = "None"
    print("Item retrieved as: " + Plate_Legs_name)
    print("\n")

    #--------------------------------------------
    # Feet
    #--------------------------------------------
    print("Feet for plate " + Plate_name)
    try:
        Plate_Feet = Plate_items['Feet']['ItemId']
        response = api.item(Plate_Feet)
        if response.ok:
            item_json = response.json()
            Plate_Feet_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Feet_name = "None"
    print("Item retrieved as: " + Plate_Feet_name)
    print("\n")

    #--------------------------------------------
    # Off Hand
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Off_Hand = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Off_Hand)
        if response.ok:
            item_json = response.json()
            Plate_Off_Hand_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Off_Hand_name = "None"
    print("Item retrieved as: " + Plate_Off_Hand_name)
    print("\n")

    #--------------------------------------------
    # Ears
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Ears = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Ears)
        if response.ok:
            item_json = response.json()
            Plate_Ears_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Ears_name = "None"
    print("Item retrieved as: " + Plate_Ears_name)
    print("\n")

    #--------------------------------------------
    # Neck
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Neck = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Neck)
        if response.ok:
            item_json = response.json()
            Plate_Neck_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Neck_name = "None"
    print("Item retrieved as: " + Plate_Neck_name)
    print("\n")

    #--------------------------------------------
    # Wrist
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Wrist = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Wrist)
        if response.ok:
            item_json = response.json()
            Plate_Wrist_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Wrist_name = "None"
    print("Item retrieved as: " + Plate_Wrist_name)
    print("\n")

    #--------------------------------------------
    # Right_Ring
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Right_Ring = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Right_Ring)
        if response.ok:
            item_json = response.json()
            Plate_Right_Ring_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Right_Ring_name = "None"
    print("Item retrieved as: " + Plate_Right_Ring_name)
    print("\n")

    #--------------------------------------------
    # Left_Ring
    #--------------------------------------------
    print("Off Hand for plate " + Plate_name)
    try:
        Plate_Left_Ring = Plate_items['Off Hand']['ItemId']
        response = api.item(Plate_Left_Ring)
        if response.ok:
            item_json = response.json()
            Plate_Left_Ring_name = item_json['item']['name']
        else:
            print(f"Failed querying item id '{item_id}' ({response.url}): [{response.status_code}] {response.reason}")
        time.sleep(wait_time)
    except:
        Plate_Left_Ring_name = "None"
    print("Item retrieved as: " + Plate_Left_Ring_name)
    print("\n")

    #------------------------------------------------
    # Create String with all items.
    #------------------------------------------------

    output_dict[i]="""
    -------------------------
    Plate Name: {0} 
    -------------------------
    Left side. 
    Plate Main Hand: {1} 
    Plate Head: {2} 
    Plate Body: {3} 
    Plate Hands: {4} 
    Plate Legs: {5} 
    Plate Feet: {6} 

    Right side. 
    Plate Off Hand: {7} 
    Plate Ear: {8} 
    Plate Neck: {9} 
    Plate Wrist: {10}
    Plate Right Ring: {11}
    Plate Left Ring: {12}

    """.format(Plate_name, Plate_Main_Hand_name, Plate_Head_name, Plate_Body_name,\
            Plate_Hands_name, Plate_Legs_name, Plate_Feet_name, Plate_Off_Hand_name,\
            Plate_Ears_name, Plate_Neck_name, Plate_Wrist_name, Plate_Right_Ring_name,\
            Plate_Left_Ring_name)

# --------------------------------------------
# Write output to file
# --------------------------------------------
for k in range(len(output_dict)):
    if k ==0:
        Print_string = output_dict[k]
    else:
        Print_string += output_dict[k]

with open(output_file, "w") as text_file:
    print(Print_string, file=text_file)


```
